```
import pandas as pd
import scipy.stats  as stats
```

```
df = pd.read_csv('https://raw.githubusercontent.com/niravjdn/Software-Measurement-Project/master/data/pit-jacoco/collection-mu-st-branch.csv', error_bad_lines=False)
```

```
import matplotlib.pyplot as plt
```

```
df.plot(x='Mutation_Score', y='Statement_Percentage', style='o')
```

```
df_clean = df.dropna()
stats.spearmanr(df_clean['Statement_Percentage'], df_clean['Mutation_Score'])
```

```
df[['Mutation_Score','Statement_Percentage']].corr(method ='spearman')
```

```
df.plot(x='Mutation_Score', y='Branch_Percentage', style='o')
```

```
df_clean = df.dropna()
stats.spearmanr(df_clean['Branch_Percentage'], df_clean['Mutation_Score'])
```

```
df[['Mutation_Score','Branch_Percentage']].corr(method ='spearman')
```

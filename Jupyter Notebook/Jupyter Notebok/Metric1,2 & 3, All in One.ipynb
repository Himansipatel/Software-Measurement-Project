```
import pandas as pd
```

```
df1 = pd.read_csv('https://raw.githubusercontent.com/niravjdn/Software-Measurement-Project/master/data/jacoco/collection/jacoco/jacoco.csv', error_bad_lines=False)
```

```
df2 = pd.read_csv('https://raw.githubusercontent.com/niravjdn/Software-Measurement-Project/master/data/jacoco/config/jacoco/jacoco.csv', error_bad_lines=False)
```

```
df3 = pd.read_csv('https://raw.githubusercontent.com/niravjdn/Software-Measurement-Project/master/data/jacoco/lang/jacoco/jacoco.csv', error_bad_lines=False)
```

```
df4 = pd.read_csv('https://raw.githubusercontent.com/niravjdn/Software-Measurement-Project/master/data/jacoco/io/jacoco/jacoco.csv', error_bad_lines=False)
```

```
df5 = pd.read_csv('https://raw.githubusercontent.com/niravjdn/Software-Measurement-Project/master/data/jacoco/math/jacoco/jacoco.csv', error_bad_lines=False)
```

```
df = pd.concat([df1, df2,df3, df4, df5], ignore_index=True)
```

```
import matplotlib.pyplot as plt
```

```
df['Statement_Percentage'] =  (df['LINE_COVERED'] / (df['LINE_COVERED'] + df['LINE_MISSED'])) * 100
```

```
df['Branch_Percentage'] =  (df['BRANCH_COVERED'] / (df['BRANCH_COVERED'] + df['BRANCH_MISSED'])) * 100
```

```
df['CC'] =  df['COMPLEXITY_COVERED'] + df['COMPLEXITY_MISSED'];
```

```
df.plot(x='CC', y='Statement_Percentage', style='o')
```

```
df.plot(x='CC', y='Branch_Percentage', style='o')
```

```
print('Average Statment Coverage '+str(df.Statement_Percentage.mean()))
```

```
print('Average Branch Coverage '+str(df.Branch_Percentage.mean()))
```
